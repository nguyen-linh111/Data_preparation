# draw sigmoid function

```python
x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))
plt.figure()
plt.plot(x,z)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
plt.show()
```

Make a line plot of the difference between the weekly maximum high price and the weekly minimum low price for Facebook. This should be a single line

```python
fb = pd.read_csv('D:\\Data preparation\\fb_stock_prices_2018.csv')
fb.head()
```

```python
fb = fb.assign(week = pd.DatetimeIndex(fb.date).week)
fb.head()
```

```python
fb_dif = fb.groupby("week").agg({"high":max,"low":min})
fb_dif.head()
```

```python
fb_different = fb_dif.high - fb_dif.low
```

```python
plt.plot(fb_different)
plt.show()
```

Draw scatter plot (kind=‘scatter’) to show the relationship between the volumn and diference between high price and low price

```python
fb_change = fb.high - fb.low
fb_volume = fb.volume
```

```python
plt.scatter(fb_change, fb_volume)
```
